```python
import pandas as pd

indices_violencia = pd.read_csv('primeiro-trimestre-2023.csv',sep=';', low_memory = False)

#pegar as colunas uteis

indices_violencia_df = indices_violencia[['Sexo_da_vítima','UF','Motivação','Faixa_etária_da_vítima','Raça_Cor_da_vítima','Profissão_da_vítima','Grupo_vulnerável']]

indeces_violencia_df = indices_violencia_df.astype('bool')

#criar variaveis para filtro

mulher_indice = indices_violencia_df['Sexo_da_vítima']=='FEMININO'
uf_indice = indices_violencia_df['UF']=='RJ'
grupo_vulneravel_indice = indices_violencia_df['Grupo_vulnerável']=='VIOLÊNCIA CONTRA A MULHER'
motivacao_indice = indices_violencia_df['Motivação']=='MOTIVAÇÃO.EM RAZÃO DE SER MULHER'
profissao_indices = indices_violencia_df['Profissão_da_vítima']=='ESTUDANTE'
preta_cor_indices = indices_violencia_df['Raça_Cor_da_vítima']=='PRETA'

#aplicar as variaveis

indices_violencia_mulher = indices_violencia_df[mulher_indice]

indices_violencia_mulher_preta = indices_violencia_df[mulher_indice & preta_cor_indices]

indices_violencia_mulher_estudante = indices_violencia_df[mulher_indice & profissao_indices]

indices_ameaca_feminicido = indices_violencia_df[mulher_indice & motivacao_indice]

indices_violencia_uf = indices_violencia_df[mulher_indice & uf_indice]

display(indices_violencia_mulher)
display(indices_violencia_mulher_preta)
display(indices_violencia_mulher_estudante)
display(indices_ameaca_feminicido)
display(indices_violencia_uf)
```

```python
#Porcentagem de Denúncia recebidas de mulheres
qntd_violencia_mulher = indices_violencia_mulher['Sexo_da_vítima'].value_counts()

qntd_violencia_total = indices_violencia_df['Sexo_da_vítima'].value_counts()

print('Dentre as denúncias recebidas no primeiro trimeste de 2023, {:.2%} foram de mulheres'.format(qntd_violencia_mulher.sum() / qntd_violencia_total.sum()))
```

```python
#Gráfico UF

qntd_violecia_uf = indices_violencia_mulher['UF'].value_counts()

grafico_violencia_uf = qntd_violecia_uf.plot(figsize = (15,5),kind ='bar', yticks = range(0,150000,15000))
grafico_violencia_uf.set_ylabel('Quantidade de denúncias')
```

```python
#Racismo

indices_violencia_mulher_negra = indices_violencia_df[mulher_indice & preta_cor_indices]

quantidade_mulheres_negras = indices_violencia_mulher_negra.value_counts().sum()

quantidade_mulheres_totais = indices_violencia_mulher.value_counts().sum()

parda_cor_indices = preta_cor_indices = indices_violencia_df['Raça_Cor_da_vítima']=='PARDA'

indices_violencia_mulher_parda = indices_violencia_df[mulher_indice & parda_cor_indices]

quantidade_mulheres_pardas = indices_violencia_mulher_parda.value_counts().sum()


quantidade_mulheres_negras_pardas = quantidade_mulheres_negras + quantidade_mulheres_pardas
#Porcentagem mulheres pretas
print('Dentre todas as denúncias feitas por mulheres, {:.2%} foram feitas por mulheres pretas'.format(quantidade_mulheres_negras/quantidade_mulheres_totais))

#Porcentagem mulheres pardas
print('Dentre todas as denúncias feitas por mulheres, {:.2%} foram feitas por mulheres pardas'.format(quantidade_mulheres_pardas/quantidade_mulheres_totais))

#Porcentagem mulheres negras e pardas
print('Dentre todas as denúncias feitas por mulheres, {:.2%} foram feitas por mulheres pretas e pardas'.format(quantidade_mulheres_negras_pardas/quantidade_mulheres_totais))
```

```python
#Porcentagem de mulheres estudantes

quantidade_mulheres_totais = indices_violencia_mulher.value_counts().sum()

quantidade_mulheres_estudantes = indices_violencia_mulher_estudante.value_counts().sum()

print('Dentre todas as denúncias feitas por mulheres, {:.2%} foram feitas por mulheres estudantes'.format(quantidade_mulheres_estudantes/quantidade_mulheres_totais))
```

```python
#Indices de ameaças de feminicídio

quantidade_mulheres_totais = indices_violencia_mulher.value_counts().sum()

quantidade_ameaca_feminicidio = indices_ameaca_feminicido.value_counts().sum()

print('Dentre todas as denúncias feitas por mulheres, {:.2%} relatam que a motivação foi somente por ela ser mulher'.format(quantidade_ameaca_feminicidio/quantidade_mulheres_totais))
```
